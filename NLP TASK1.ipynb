```python
import re

# Basic pattern matching
pattern1 = re.compile(r'abc')
result1 = pattern1.match('abcdef')
if result1:
    print("Match found at the beginning:", result1.group())

# Searching for a pattern anywhere in the string
pattern2 = re.compile(r'l')
result2 = pattern2.search('hello')
if result2:
    print("Character 'l' found:", result2.group())

# Finding all occurrences of a pattern
pattern3 = re.compile(r'\d+') # one or more digits
result3 = pattern3.findall('There are 123 apples and 45 oranges.')
print("Numbers found:", result3)

# Corrected pattern for social security number and variable usage
pattern4 = re.compile(r'\d{3}_\d{2}_\d{4}')
result4 = pattern4.match('123_45_6789')
if result4:
  print("Social security number:",result4.group())

# Corrected pattern for .com and variable usage
pattern5 = re.compile(r'\.com')
result5 = pattern5.search("www.example.com")
if result5:
  print(".com found in",result5.group())

# Corrected pattern for date and variable usage
pattern6 = re.compile(r'(\d{2})/(\d{2})/(\d{4})')
result6 = pattern6.match('09/09/2024')
if result6:
  print("Day:",result6.group(1))
  print("Month:",result6.group(2))
  print("Year:",result6.group(3))

# Corrected pattern for "cat" or "dog" and variable usage
pattern7 = re.compile(r'cat|dog')
result7 = pattern7.search('i have a cat and a dog')
if result7:
  print("Animal found:",result7.group())
```
